The final part to this section is plotting code. The first part of the plotting code below

* sends the tut_power_plot DataFrame into a VegaLite pipeline using |>.

* @vlplot() initializes a base VegaLite plot 

* '+' is used to layer multiple visual elements (charts) on top of one another.

The next part creates two plots, one area plot and one line plot

```julia
using VegaLite
```

```julia
tut_power_plot  |>
@vlplot()+
@vlplot(mark={:area},
    x={:Hour,title="Time Step (hours)",labels="Resource_Type:n",axis={values=0:12:168}}, y={:MW,title="Load (MW)",type="quantitative"},
    color={"Resource_Type:n",scale={scheme="accent"},sort="descending"},order={field="Resource_Type:n"},width=845,height=400)+
@vlplot(mark=:line,x=:Hour,y=:Demand_Total,lables="Demand",color={datum="Demand",legend={title=nothing}},title="Resource Capacity per Hour with Load Demand Curve, all Zones")
```
